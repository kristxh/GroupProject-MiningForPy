#### Called Crypto Compare API to get price data for BTC, XRP, ETH, LTC, EOS, DASH

```python
# Save url and config information
url = "https://min-api.cryptocompare.com/data/histoday?fsym="
currency = "USD"
limit = 1460
symbol = "DASH"

# Build partial query URL
query_url = f"{url}{symbol}&tsym={currency}&limit={limit}&api_key={api_key}"
```

```python
response = requests.get(query_url)
response_json = response.json()
print(json.dumps(response_json, indent=4, sort_keys=True))
```

```python
prices = requests.get(query_url).json()
df = pd.DataFrame.from_dict(json_normalize(prices["Data"]), orient='columns')
df.to_csv("Resources/crypto-dash.csv", header=True)
df.head()
```
