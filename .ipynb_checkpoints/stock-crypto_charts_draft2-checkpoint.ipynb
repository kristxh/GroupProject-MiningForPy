```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
```

```python
#Load stock and crypto price files
stock_data = "Resources/all_stock_data.csv"
crypto_data = "Resources/all_crypto_data.csv"

# Create stock price DataFrames
stock_df = pd.read_csv(stock_data)
stock_df['date'] =  pd.to_datetime(stock_df['date'])
crypto_df = pd.read_csv(crypto_data).sort_values('date').reset_index(drop=True)
crypto_df['date'] =  pd.to_datetime(crypto_df['date'])
```

```python
# Create a chart of selected cryptos
fig, ax1 = plt.subplots()
ax1.set_title('Daily Cryptocurrency Prices')
color = 'tab:green'
ax1.set_ylabel('Price BTC (USD)', color=color)
ax1.plot(crypto_df['date'], crypto_df['close_btc'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
plt.xticks(rotation=50)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.grid(axis='both')
ax2.set_ylabel('Price Other Cryptos (USD)', color=color)  
ax2.plot(crypto_df['date'], crypto_df['close_xrp'],color='coral',label="XRP")
ax2.plot(crypto_df['date'], crypto_df['close_eth'],color='coral', label="ETH")
ax2.plot(crypto_df['date'], crypto_df['close_ltc'],color='indigo', label="LTC")
ax2.plot(crypto_df['date'], crypto_df['close_eos'],color='orchid', label="EOS")
ax2.plot(crypto_df['date'], crypto_df['close_dash'],color=color,label="Dash")
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper left')

fig.tight_layout() 
plt.show()
```

```python
# Create a chart of selected stocks
fig, ax = plt.subplots()
ax.grid(axis='y')
ax.set_title('Daily Stock Prices')
ax.set_ylabel('Price (USD)')
ax.plot(stock_df['date'], stock_df['close_amd'], color='green', label='AMD')
ax.plot(stock_df['date'], stock_df['close_nvda'], color='orange',label='NVDA')
ax.plot(stock_df['date'], stock_df['close_mu'], color='blue',label='MU')
ax.tick_params(axis='y')
plt.xticks(rotation=50)
ax.legend(loc='upper left')
```

```python
# Compare Bitcoin with Stock Prices

fig, ax1 = plt.subplots()

color = 'tab:green'
# ax1.set_xlabel('time (s)')
ax1.set_ylabel('Price BTC (USD)', color=color)
ax1.plot(crypto_df['date'], crypto_df['close_btc'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
# ax1.xaxis.set_major_locator(years)
fig.autofmt_xdate()

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Price NVDA (USD)', color=color)  # we already handled the x-label with ax1
ax2.plot(stock_df['date'], stock_df['close_nvda'],color=color)
ax2.plot(stock_df['date'], stock_df['close_mu'],color='coral')
ax2.plot(stock_df['date'], stock_df['close_amd'],color='orange')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
