```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
#Load gpu price files
gpu_data = "Resources/all_gpu_price.csv"
gpu_df = pd.read_csv(gpu_data).set_index('date')

crypto_data = "Resources/all_crypto_data.csv"
crypto_data_df = pd.read_csv(crypto_data)
crypto_data_df['date'] =  pd.to_datetime(crypto_data_df['date'])
crypto_data_df.set_index('date')
crypto_data_df['weekday'] = crypto_data_df['date'].dt.weekday
crypto_data_df.head(10)
```

```python
#Creating Data Frames Per GPU Type
df_1080_ti = pd.DataFrame({"Nvidia 1080 Ti": gpu_df['Nvidia_1080_Ti'],
                       "ASUS 1080 Ti": gpu_df['ASUS_1080_TI'],
                       "Gigabyte 1080 Ti":gpu_df['Gigabyte_1080_Ti']}).dropna(how='all')
#Need to set index format same as crypto data df
df_1080_ti.index =  pd.to_datetime(df_1080_ti.index)
```

```python
#Merge DF's With Crypto Data
merge_1080_ti = pd.merge(df_1080_ti, crypto_data_df, on="date", how='left').set_index('date')
#Making new DF with just data from same weekday to produce weekly data
only_mondays_df = merge_1080_ti[merge_1080_ti['weekday'] == 1]
#calculating pct chg for weekly data
pct_chg_df = only_mondays_df.pct_change()
pct_chg_df.head(1)
```

```python
pltcol = ['Nvidia 1080 Ti',
          'ASUS 1080 Ti',
          'Gigabyte 1080 Ti',
          'close_btc',
          'close_xrp',
          'close_eth',
          'close_ltc',
          'close_eos',
          'close_dash']

plt.figure(figsize=(14,10))
plt.plot(pct_chg_df[pltcol])
plt.colors()
plt.legend(pct_chg_df[pltcol], loc='upper right')
plt.savefig('All GPU % Change vs Crypto % Change')
```

```python
gpus =['Nvidia 1080 Ti',
      'ASUS 1080 Ti',
      'Gigabyte 1080 Ti'
     ]
cryptos = ['close_btc',
           'close_xrp',
           'close_eth',
           'close_ltc',
           'close_eos',
           'close_dash'
          ]

```

```python
plt.figure(figsize=(20,10))
for i in range(0, 6):
    plt.suptitle('WoW - Crypto Currency Closing Price Change (%)')
    plt.subplot(2, 3, i + 1).plot(pct_chg_df[cryptos[i]])
    plt.title(cryptos[i])
    plt.xticks(rotation=45)
    plt.ylabel('% Change')
    plt.xlabel('Week of')
    plt.subplots_adjust(hspace=.5)
plt.savefig('All Crypto % Change')
```

```python
plt.figure(figsize=(20,10))
for i in range(0, 3):
    plt.suptitle('WoW - GPU Sales Price Change(%)')
    plt.subplot(1, 3, i + 1).plot(pct_chg_df[gpus[i]])
    plt.title(gpus[i])
    plt.xticks(rotation=45)
    plt.ylabel('% Change')
    plt.xlabel('Week of')
    plt.subplots_adjust(hspace=.5)
plt.savefig('All GPU % Change')
```
