```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
#Load stock and crypto price files
stock_data = "Resources/all_stock_data.csv"
crypto_data = "Resources/all_crypto_data.csv"

# Create stock price DataFrames
stock_df = pd.read_csv(stock_data)
stock_df['date'] =  pd.to_datetime(stock_df['date'])
crypto_df = pd.read_csv(crypto_data) 
crypto_df['date'] =  pd.to_datetime(crypto_df['date'])
```

### Daily Price Comparison Across Cryptos
* All cryptos followed a similar increase and decrease trend from Jan 2017 - Dec 2018.
* The peak was in Dec 2017

```python
# Create a chart of selected cryptos
fig, ax1 = plt.subplots()
ax1.set_title('Daily Cryptocurrency Prices (USD)')
color = 'tab:blue'
ax1.set_ylabel('Price BTC (USD)', color=color)
ax1.plot(crypto_df['date'], crypto_df['close_btc'], color=color,label="BTC")
ax1.tick_params(axis='y', labelcolor=color)
plt.xticks(rotation=50)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:green'
ax2.grid(axis='both')
ax2.set_ylabel('Price Other Cryptos (USD)', color=color)  
ax2.plot(crypto_df['date'], crypto_df['close_xrp'],color='lightgreen',label="XRP")
ax2.plot(crypto_df['date'], crypto_df['close_eth'],color='coral', label="ETH")
ax2.plot(crypto_df['date'], crypto_df['close_ltc'],color='darkblue', label="LTC")
ax2.plot(crypto_df['date'], crypto_df['close_eos'],color='orchid', label="EOS")
ax2.plot(crypto_df['date'], crypto_df['close_dash'],color=color,label="Dash")
ax2.tick_params(axis='y', labelcolor=color)
fig.legend(bbox_to_anchor=(1.2, 1.0))
fig.tight_layout() 
fig.savefig("Images/daily_prices_crypto.png")
```

### Daily Price Comparison Across Stocks
* All three stocks saw price growth beyond the Dec 2017 peak of crypto prices.
* Micron has been on the decline since June 2018.
* Invidia and AMD have been declining since Oct 2018.

```python
# Create a chart of selected stocks
fig, ax = plt.subplots()
ax.grid(axis='y')
ax.set_title('Daily Stock Prices(USD)')
ax.set_ylabel('Price (USD)')
ax.plot(stock_df['date'], stock_df['close_amd'], color='green', label='AMD')
ax.plot(stock_df['date'], stock_df['close_nvda'], color='orange',label='NVDA')
ax.plot(stock_df['date'], stock_df['close_mu'], color='lightblue',label='MU')
ax.tick_params(axis='y')
plt.xticks(rotation=50)
fig.legend(bbox_to_anchor=(1.2, 1.0))
fig.tight_layout() 
fig.savefig("Images/daily_prices_stock.png")
```

### Daily Price Comparison - BTC and Stocks
* BTC peaks in Dec 2018 while the stocks continue to rise until Oct 2018.
* BTC continues to fall along with stock prices after Oct 2018.

```python
# Compare Bitcoin with Stock Prices
fig, ax1 = plt.subplots()

color = 'tab:blue'
# ax1.set_xlabel('time (s)')
ax1.set_ylabel('Price BTC (USD)', color=color)
ax1.plot(crypto_df['date'], crypto_df['close_btc'], color='darkblue', label="Crypto - BTC")
ax1.tick_params(axis='y', labelcolor=color)
fig.autofmt_xdate()

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:green'
ax2.grid(axis='both')
ax2.set_ylabel('Stock Prices for Video Card Manufacturers (USD)', color=color)  
ax2.plot(stock_df['date'], stock_df['close_nvda'],color='orange',label="Stock - NVDA")
ax2.plot(stock_df['date'], stock_df['close_mu'],color='lightblue',label="Stock - MU")
ax2.plot(stock_df['date'], stock_df['close_amd'],color='green',label="Stock - AMD")
ax2.tick_params(axis='y', labelcolor=color)
fig.legend(bbox_to_anchor=(1.3, 1.0))
fig.tight_layout()
fig.savefig("Images/daily_prices_btc_stocks.png")
```

### Daily Price Comparison - ETH and Stocks
* ETH follows a similar pattern to BTC, but peaks slightly later in Jan 2018 before it begins a significant drop.
