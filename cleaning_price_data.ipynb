```python
# Dependencies and Setup
import pandas as pd
from datetime import datetime
```

```python
#Load stock price files
amd_data = "Resources/stock-amd.csv"
nvda_data = "Resources/stock-nvda.csv"
mu_data = "Resources/stock-mu.csv"

# Load crypto price files - (BTC, XRP, ETH, LTC, EOS, DASH)
btc_data = "Resources/crypto-btc.csv"
xrp_data = "Resources/crypto-xrp.csv"
eth_data = "Resources/crypto-eth.csv"
ltc_data = "Resources/crypto-ltc.csv"
eos_data = "Resources/crypto-eos.csv"
dash_data = "Resources/crypto-dash.csv"

# Create stock price DataFrames
amd_df = pd.read_csv(amd_data)
nvda_df = pd.read_csv(nvda_data)
mu_df = pd.read_csv(mu_data)

# Create crypto price DataFrames
btc_df = pd.read_csv(btc_data)
xrp_df = pd.read_csv(xrp_data)
eth_df = pd.read_csv(eth_data)
ltc_df = pd.read_csv(ltc_data)
eos_df = pd.read_csv(eos_data)
dash_df = pd.read_csv(dash_data)
```

```python
# Display stock price DataFrame
amd_df.head()
```

```python
# Change the date format and create a new DataFrame with only date and closing price columns
clean_amd_df = amd_df[['date','close']].set_index('date', drop=True)
clean_amd_df = clean_amd_df[clean_amd_df.index != '16:00']
clean_amd_df.index = pd.to_datetime(clean_amd_df.index, format="%m/%d/%Y")

# Export to csv
clean_amd_df.to_csv("Resources/cl_stock-amd.csv", header=True)

# Display clean data
clean_amd_df.head()
```
